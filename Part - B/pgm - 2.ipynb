```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'tips.csv'
data = pd.read_csv(path)
```

```python
data.head(10)
```

```python
sns.scatterplot(x='day', y='tip', data=data)
plt.title('Scatter plot')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
sns.lineplot(x='total_bill', y='tip', data=data)
plt.title('Scatter plot')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.hist(data['total_bill'], edgecolor='black', bins=10)
plt.title('Histogram')
plt.xlabel('Total bill')
plt.ylabel('Frequency')
plt.show()
```
